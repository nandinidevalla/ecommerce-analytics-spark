Q4.	Calculate total revenue, average order value, and monthly revenue trends.

a. Toal revenue

```
%sql
SELECT sum(payment_value) as Total_Revenue
FROM payments_view
```

Average order Value

```
%sql
SELECT count(DISTINCT(order_id)), count(order_id) 
FROM payments_view
```

```
%sql
SELECT order_id, count(order_id) as c
FROM payments_view
group by order_id
order by c desc
limit 5
```

we see that number of order_ids is not unique. Hence we need to first find out the total value of an order_id and then average it

```
%sql
select avg(sales) 
from (
  SELECT sum(payment_value) as sales
  FROM payments_view
  group by order_id 
) as tab1
```

Monthly Trends

```
%sql
SELECT DATE_FORMAT(o.order_purchase_timestamp, 'yyyy-MM') AS year_month , sum(sales) as monthly_revenue
FROM orders_view as o
LEFT JOIN (
  SELECT order_id, sum(payment_value) as sales
  FROM payments_view
  group by order_id  
  ) AS tab1
on o.order_id=tab1.order_id
GROUP BY year_month
ORDER BY year_month asc

```

Q5.	Identify the top 5 best-selling product categories.

```
%sql
SELECT p.product_category_name, sum(total_orders) AS TOT_ORDERS
FROM products_view as p
left join (
  SELECT product_id, count(order_id) as total_orders
FROM op_view
GROUP BY product_id
) as tab1
on p.product_id=tab1.product_id
GROUP BY p.product_category_name
ORDER BY tot_orders DESC
LIMIT 5
```

Q13.	Analyze the distribution of review scores.

```
%sql
select review_Score from reviews_view
```

```
%sql
SELECT review_score, COUNT(*) AS review_count, concat(round(count(*)/(select count(*) from reviews_view where review_score IN ('1', '2', '3', '4', '5'))*100,2), '%') as percentage_distribution
FROM reviews_view
WHERE review_score IN ('1', '2', '3', '4', '5')
GROUP BY review_score
ORDER BY review_score;
```

Q14.	Find the percentage of orders with low (1-2) vs. high (4-5) reviews.

```
%sql
SELECT 
    CASE 
        WHEN review_score IN ('1', '2') THEN 'Low Reviews (1-2)'
        WHEN review_score IN ('4', '5') THEN 'High Reviews (4-5)'
        WHEN review_score IN ('3') THEN 'Mid Reviews (3)'
    END AS review_category,
    COUNT(*) AS review_count,
    CONCAT(ROUND(COUNT(*) / (SELECT COUNT(*) FROM reviews_view WHERE review_score IN ('1', '2', '3', '4', '5')) * 100, 2), '%') AS percentage_distribution
FROM reviews_view
WHERE review_score IN ('1', '2', '3', '4', '5')
GROUP BY review_category
ORDER BY review_category;
```

Q16.	Find the month with the highest sales in 2018.

```
%sql
SELECT DATE_FORMAT(o.order_purchase_timestamp, 'yyyy-MM') AS year_month , sum(sales) as monthly_revenue
FROM orders_view as o
LEFT JOIN (
  SELECT order_id, sum(payment_value) as sales
  FROM payments_view
  group by order_id  
  ) AS tab1
on o.order_id=tab1.order_id
WHERE DATE_FORMAT(o.order_purchase_timestamp, 'yyyy') = 2018
GROUP BY year_month
ORDER BY monthly_revenue DESC

```

Q18.	Determine whether higher review scores correlate with faster delivery times.

```
%sql
SELECT 
    r.review_score,
    AVG(DATEDIFF(o.order_delivered_customer_date, o.order_purchase_timestamp)) AS avg_delivery_time
FROM 
    orders_view o
JOIN 
    reviews_view r ON o.order_id = r.order_id
WHERE 
    o.order_delivered_customer_date IS NOT NULL
    AND o.order_purchase_timestamp IS NOT NULL
    AND r.review_score IS NOT NULL
GROUP BY 
    r.review_score
ORDER BY 
    r.review_score;
```

Q19.	Identify orders with suspiciously high transaction values.

```
WITH order_payment_totals AS (
    SELECT 
        p.order_id, 
        SUM(p.payment_value) AS total_payment_value
    FROM 
        payments_view p
    GROUP BY 
        p.order_id
)
SELECT 
    opt.order_id,
    opt.total_payment_value
FROM 
    order_payment_totals opt
WHERE 
    opt.total_payment_value > (SELECT PERCENTILE_CONT(0.95) WITHIN GROUP (ORDER BY total_payment_value) FROM order_payment_totals)
ORDER BY 
    opt.total_payment_value DESC;
```
